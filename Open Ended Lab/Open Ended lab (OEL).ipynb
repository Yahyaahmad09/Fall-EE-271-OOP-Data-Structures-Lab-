# <span style="color: blue; font-size: 1em;"> Step 2: Dataset Preparation (Splitting and Scaling)</span>
1) Data is divided into the **`Train set`** and **`Test set`**.
2) We use the `Train set` to make the algorithm learn the `data‚Äôs` behavior and then check the accuracy of our model on the `Test set`.

##### ***Features (X):*** 
     The columns that are inserted into our model will be used to make predictions. 
##### ***Prediction (y):*** 
     Target variable that will be predicted by the features



```python
data.dropna(inplace=True)
```

#### <span style="color: blue; font-size: 1em;"> Remove the price column from the Data</span>

```python
X = data.drop('price',axis =1).values
```

#### <span style="color: blue; font-size: 1em;"> Get the price column from the Data</span>

```python
y = data['price'].values
```

#### <span style="color: blue; font-size: 1em;"> Spliting Data into Train and Test </span>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

#### <span style="color: blue; font-size: 1em;"> Scale the Data ito Train and Test </span>

```python
s_scaler = StandardScaler()
X_train = s_scaler.fit_transform(X_train)
X_test = s_scaler.transform(X_test)
```

#### <span style="color: green; font-size: 1.5em;"> Step 3: Model Selection and Evaluation </span>
 #### <span style="color: blue; font-size: 1.25em;"> üí°Model 1: Multiple Linear Regressions</span>
 Multiple Linear Regressions **`Multiple Linear Regression`** is an extension of *`Simple Linear Regression`* and assume that there is a linear relationship between a dependent variable `Y` and independent variables `X`

Let‚Äôs wrap the training process in our Regression model:

#### <span style="color: green; font-size: 1em;"> Multiple Regerssion </span>

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

#### <span style="color: green; font-size: 1em;"> Checking the Model Coefficients </span>

```python
print(regressor.intercept_)
print(regressor.coef_)
```

#### <span style="color: green; font-size: 1.5em;"> Prediction </span>

```python
y_pred = regressor.predict(X_test)
```

#### <span style="color: green; font-size: 1em;"> Convert the Coefficient into a DataForm </span>

```python
coeff_df = pd.DataFrame(regressor.coef_, data.drop('price',axis =1).columns, columns=['Coefficient']) 
coeff_df
```

#### <span style="color: green; font-size: 1.1em;"> Visualizating Y_test and Predict </span>

```python
fig = plt.figure(figsize=(10,5))
residuals = (y_test- y_pred)
sns.distplot(residuals)
```

#### <span style="color: green; font-size: 1.2em;"> Compare the Actual and Predicted Values </span>
Let‚Äôs compare **`actual output`** and **`predicted value`** to measure how far our predictions are from the `real house prices`

```python
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = df.head(10)
df1
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```

#### <span style="color: blue; font-size: 1.25em;"> üí°Model 2: Keras Regressions </span>
Let‚Äôs create a baseline **`neural network model`** for the regression problem. Starting with all of the needed functions and objects.



#### <span style="color: blue; font-size: 1em;"> Creating a Sequential Model </span>
- Then, we train the model for `400 epochs`, and each time record the training and `validation accuracy` in the history object.
- To keep track of how well the model is performing for each epoch, the model will run in both `train` and `test` data along with calculating the `loss function`.


```python
model = Sequential()
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='Adam',loss='mes')
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

#### <span style="color: blue; font-size: 1em;"> Train Neural Network </span>
- Summary

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
model.summary()
```

#### <span style="color: blue; font-size: 1.2em;"> Visualizing the losses </span>

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot(figsize=(12,8))
```

#### <span style="color: blue; font-size: 1em;"> Evaluation on Test Data¬∂ </span>

```python
y_pred = model.predict(X_test)

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```
